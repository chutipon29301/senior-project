```python
import requests 
import json
# defining the api-endpoint  
API_ENDPOINT = " http://smartcontract.example.com:5000"
# data to be sent to api
data = {"value":"ming"} 
# sending post request and saving response as response object 
r = requests.post(url = API_ENDPOINT, json = data) 
# extracting response text  
result = json.loads(r.text)
print("The pastebin URL is: %s"%result['value'])
```

```python
import pandas as pd
df=pd.read_csv('data/data.csv')
df
```

```python
# data = {"value":"ming"} 
import datetime
hr_df=df.loc[df['Time'] == '08:00']
buyer_name=['CHAM1','CHAM2','CHAM3','CHAM4','CHAM5']
buyer_price=[3.9,3.2,4.7,3.8,3.3] #[4,3,2,1,5]
buyer_q=[float(hr_df[name]) for name in buyer_name]
seller_name=['CHAM1-PV','CHAM2-PV','CHAM3-PV','CHAM4-PV','CHAM5-PV']
seller_price=[5.0,4.9,3.5,3.1,3.7] #[2,3,4,5,1]
seller_q=[float(hr_df[name]) for name in seller_name]
buyers=[{'quantity':buyer_q[i],'bidPrice':buyer_price[i],'timestamp':datetime.datetime.now().isoformat()} for i in range(len(buyer_name))]
sellers=[{'quantity':seller_q[j],'bidPrice':seller_price[j],'timestamp':datetime.datetime.now().isoformat()} for j in range(len(seller_name))]

data = {"buyers":buyers,"sellers":sellers}
print(data)
```

```python
# url=API_ENDPOINT+'/uniKDA'
url=API_ENDPOINT+'/disKDA'
# url=API_ENDPOINT+'/weightedAvg'

r = requests.post(url = url, json = data) 
# extracting response text 
result = json.loads(r.text)
result
```
